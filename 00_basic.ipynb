# 2. Basic file operation

You can read the file in two ways. 

1. Open a ROOT file directly
2. Using the `event` class

The `event` class provides a few helper functions that facilitates your work flow. For example:

**hits**
* get_hits_truth(): return all the hits information (truth)
* get_hits_digi(): return all the hits information (digitized)

**tracks**
* get_truthtrack(): return all the truth tracks
* get_recon_kalman(): return all the Kalman filter reconstructions

### 2.1 Open a ROOT file directly


```python
# 1. filename
muon_energy = 100 #GeV
data_dir    = f"/project/def-mdiamond/tomren/mathusla/data/fit_study_6layer/muon_{muon_energy}_GeV"
files=glob.glob(f"{data_dir}/*/*/stat*.root",)
files.sort(key=os.path.getmtime)

filename = files[-1]
# filename='../../simdata/fit_study/run0.root'
print(files)
print(filename,"\n")

# 2. Open file and print branches
tfile = root.TFile.Open(filename)
tree_name = ["integral_tree","box_run"][0]
tree_name = tfile.GetListOfKeys()[0].GetName()
Tree = tfile.Get(tree_name)

barnches = Tree.GetListOfBranches()
nevents = int(Tree.GetEntries())
branch_list = [barnches[i].GetName() for i in range(len(barnches))]
print(branch_list)

# Basic file operation:
# First, use Tree.GetEntry(i_th_entry) to set the current entry to i_th_entry
Tree.GetEntry(0)
# Then, you can access the content of any brach of this tree at i_th entry with dot method: Tree.BRANCH_NAME
print("\nHit_x:", Tree.Hit_x)
```

# 3. Visualization

Like the file operation, there are two ways to visualize an event

1. Manually plot the x,y,z of all hits
2. Using pre-defined functions in `visualizaion`

### 3.1 Manually plot the x,y,z of all hits

```python
ev = event.Event(filename, 0, tree_name=tree_name)
for i in range(300):
    ev.EventNumber=i

    tracks=ev.get_truthtrack()
    if len(tracks)>0:
        for track in tracks:
            track=util.coord_cms2det(track)
            # plt.plot(track[0],track[2],marker=".",color="grey",alpha=0.1)
            plt.plot(track[0],track[2],marker=".",color="C1",alpha=0.1)    
        
    plt.xlabel('x [m]')
    plt.ylabel('z [m]')
    
vis.drawdet_xz()
```

### 3.2 Using pre-defined functions in `visualizaion`

Plot truth, digitization, reconstruction of one event

**Truth, track level:** vis.plot_truth():

**Truth, hit level:**  vis.plot_multiple_events(): plot the low-level truth of multiple events

**Digi, hit level:** vis.plot_digi()

**Reconstruction:**  vis.plot_recon()

```python
ev.EventNumber=25

# Plot truth+digi
fig,axs=plt.subplots(2,2,figsize=(12,9))
fig = vis.plot_truth(ev,fig=fig, disp_det_view=False, disp_vertex=False, disp_first_hit=False);
fig = vis.plot_digi(ev,fig=fig, disp_det_view=True);
plt.show()


# Plot truth+recon
fig,axs=plt.subplots(2,2,figsize=(12,9))
# fig = vis.plot_truth(ev,fig=fig, disp_det_view=False, disp_vertex=False, disp_first_hit=False);
fig = vis.plot_recon(ev,fig=fig, disp_det_view=True, disp_recon_vertex=False);
plt.show()
```
